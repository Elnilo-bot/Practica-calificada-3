```python

import requests
import re
import pandas as pd

# ── Parte II: Recuperación de datos ──────────────────────────────────────────

URL = "https://openlibrary.org/search.json"
params = {
    "subject": "fiction",   # tema de búsqueda
    "limit": 20,            # al menos 15 registros
    "fields": "title,author_name,first_publish_year,isbn,number_of_pages_median"
}

response = requests.get(URL, params=params)
print("Status code:", response.status_code)          # 200 = éxito
print("Content-Type:", response.headers["Content-Type"])

data = response.json()
libros_raw = data["docs"]
print(f"\nRegistros obtenidos: {len(libros_raw)}")
```

```python
# ── Extracción de 3 atributos relevantes ─────────────────────────────────────
# 1. título, 2. autor, 3. año de primera publicación

registros = []
for libro in libros_raw:
    registros.append({
        "titulo":   libro.get("title", "Desconocido"),
        "autor":    libro.get("author_name", ["Desconocido"])[0],
        "anio":     libro.get("first_publish_year", None),
        "isbn":     libro.get("isbn", [None])[0],       # para la regex
        "paginas":  libro.get("number_of_pages_median", None)
    })
```

```python
# ── Expresión regular: validar formato ISBN-13 ────────────────────────────────
patron_isbn13 = re.compile(r"^\d{13}$")

for r in registros:
    isbn = str(r["isbn"]) if r["isbn"] else ""
    r["isbn13_valido"] = bool(patron_isbn13.match(isbn))

# Ejemplo: mostrar los que tienen ISBN-13 válido
for r in registros:
    if r["isbn13_valido"]:
        print(f"✔ ISBN-13 válido: {r['isbn']}  →  {r['titulo']}")
```

```python
import requests
import pandas as pd

# -----------------------------
# Obtener datos desde Open Library
# -----------------------------
url = "https://openlibrary.org/search.json"

params = {
    "q": "programming",
    "limit": 50
}

response = requests.get(url, params=params)
datos = response.json()

libros = datos["docs"]

# -----------------------------
# Crear el DataFrame
# -----------------------------
lista = []

for libro in libros:
    lista.append({
        "Titulo": libro.get("title"),
        "Autor": ", ".join(libro.get("author_name", []))
                  if libro.get("author_name") else "Desconocido",
        "Año": libro.get("first_publish_year"),
        "Ediciones": libro.get("edition_count", 0),
        "Idioma": libro.get("language", ["Desconocido"])[0]
                  if libro.get("language") else "Desconocido"
    })

df = pd.DataFrame(lista)

print("Primeras filas del DataFrame:")
print(df.head())

# -----------------------------
# Limpieza de datos
# -----------------------------

# Eliminar registros sin año
df = df.dropna(subset=["Año"])

# Convertir Año a entero
df["Año"] = df["Año"].astype(int)

# Eliminar títulos duplicados
df = df.drop_duplicates(subset="Titulo")

# -----------------------------
# Estadísticas descriptivas
# -----------------------------

print("\nCantidad total de libros:", len(df))

print("\nPromedio del año de publicación:")
print(df["Año"].mean())

print("\nLibro con mayor número de ediciones:")
print(df["Ediciones"].max())

print("\nCinco idiomas más frecuentes:")
print(df["Idioma"].value_counts().head())
# parte hecha por Mateo
```

```python
#PARTE 4 - BARRANTES ELKIN

import matplotlib.pyplot as plt

# Top 10 autores más frecuentes
top_autores = df["autor"].value_counts().head(10)

plt.figure(figsize=(10,5))

top_autores.plot(kind="bar")

plt.title("Top 10 autores más frecuentes")
plt.xlabel("Autor")
plt.ylabel("Cantidad de libros")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()


"""
El gráfico muestra los autores con mayor cantidad de libros dentro de la
muestra obtenida desde Open Library. Los autores con barras más altas tienen
una mayor representación en el conjunto de datos analizado.
```
