```python
# PARTE I : PROGRAMAS EN RED

import requests

url = "https://openlibrary.org/search.json?q=python"

respuesta = requests.get(url)

print("Status Code:", respuesta.status_code)
print("URL consultada:", respuesta.url)
print("Content-Type:", respuesta.headers.get("Content-Type"))

if respuesta.status_code == 200:
    print("La conexión fue exitosa.")
else:
    print("La conexión falló.")
```
